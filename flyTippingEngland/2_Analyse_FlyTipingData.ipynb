```python
total_incidents=df['total'].sum()
london_incidents=df[df['region']=='London']['total'].sum()
london_incidents_perc=london_incidents/total_incidents*100

print('total_incidents: ', total_incidents/7)
print('london_incidents: ', london_incidents)
print('london_incidents_perc: ', london_incidents_perc)
```

```python
total_clearance_costs=df['total_clearance_costs'].sum()
london_clearance_costs=df[df['region']=='London']['total_clearance_costs'].sum()
london_clearance_costs_perc=london_clearance_costs/total_clearance_costs*100

print('total_clearance_costs: ', total_clearance_costs/7)
print('london_clearance_costs: ', london_clearance_costs)
print('london_clearance_costs_perc: ', london_clearance_costs_perc)
```

```python
total_incidents=df['total'].sum()
highway_incidents=df['highway'].sum()
highway_incidents_perc=highway_incidents/total_incidents*100

print('total_incidents: ', total_incidents)
print('highway_incidents: ', highway_incidents)
print('highway_incidents_perc: ', highway_incidents_perc)
```

```python
total_incidents=df['total'].sum()
other_household_waste=df['other_household_waste'].sum()
other_household_waste_perc=other_household_waste/total_incidents*100

print('total_incidents: ', total_incidents)
print('other_household_waste: ', other_household_waste)
print('other_household_waste_perc: ', other_household_waste_perc)
```

```python
#Focus on reducing London’s highway and other_household_waste fly-tipping incidents
total_incidents=df['total'].sum()
london_highway_otherwaste=df[df['region']=='London']['highway'].sum() + df[df['region']=='London']['other_household_waste'].sum()
london_highway_otherwaste_perc=london_highway_otherwaste/total_incidents*100

print('total_incidents: ', total_incidents)
print('london_highway_otherwaste: ', london_highway_otherwaste)
print('london_highway_otherwaste_perc: ', london_highway_otherwaste_perc)
```

```python
#split df in categories
df_loc=df.iloc[:,[0,1,2,3,4]+list(range(5,15))]
df_type=df.iloc[:,[0,1,2,3,4]+list(range(15,30))]
df_cost_type=df.iloc[:,[0,1,2,3,4]+list(range(30,44,2))]
df_cost=df.iloc[:,[0,1,2,3,44]+list(range(31,45,2))]
df_cost.rename({'total_clearance_costs':'total'}, axis=1,inplace=True)
```

```python
df_all_name=[(df_loc,'waste_location_count'),(df_type,'waste_type_count'),
             (df_cost_type,'waste_removal_count'),(df_cost,'waste_removal_cost_count')]

for df,name in df_all_name:
    display(name, df.shape)
    display(df.head(n))
```

```python
for df,name in df_all_name:
    display(name,df.sort_values('total', ascending=False).head(n))
```

```python
for df,name in df_all_name:
    display(name,df.groupby(['region']).sum())
```

```python
for df,name in df_all_name:
    display(name,df.groupby(['year']).sum())
```
